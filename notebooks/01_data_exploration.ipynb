```python
import pandas as pd



# Only load columns we need — ignore heavy NLP columns
cols_to_keep = ['id', 'author', 'date', 'score', 'num_comments', 
                'title', 'compound', 'neg', 'neu', 'pos']

gme = pd.read_csv("/Users/annu/GME_CIB_Detector/data/rGME_dataset_features.csv",
                  usecols=lambda c: c in cols_to_keep)

jungle = pd.read_csv("/Users/annu/GME_CIB_Detector/data/rGMEJungle_dataset_features.csv",
                     usecols=lambda c: c in cols_to_keep)

print(f"GME rows: {len(gme)}, columns: {list(gme.columns)}")
print(f"Jungle rows: {len(jungle)}, columns: {list(jungle.columns)}")
```

```python
gme['date'] = pd.to_datetime(gme['date']) 
jungle['date'] = pd.to_datetime(jungle['date'])

#checking date range
print("---GME date range---")
print(f"from : {gme['date'].min()}")
print(f"to : {gme['date'].max()}")

print("---GMEJungle date range---")
print(f"from : {jungle['date'].min()}")
print(f"to : {jungle['date'].max()}")

```

```python
#period in which the hyping of gme and prices started to rise
squeeze = gme[(gme['date'] >= '2020-12-01') & (gme['date'] <= '2021-01-28')]

print(f"Total r/GME posts: {len(gme)}")
print(f"Posts in extended squeeze window: {len(squeeze)}")
print(f"\nPosts per day:")
print(squeeze.groupby('date')['id'].count())
```

```python
# Cell 4: EDA on squeeze window

# Convert date column first
gme['date'] = pd.to_datetime(gme['date'])

# Now filter
squeeze = gme[(gme['date'] >= '2021-01-04') & (gme['date'] <= '2021-01-28')].copy()

print(f"no of posts in squeeze period : {len(squeeze)}")
print(f"Date range: {squeeze['date'].min()} to {squeeze['date'].max()}")
print(f"unique authors: {squeeze['author'].nunique()}")

# Top 20 authors
print("\n top 20 author:")
print(squeeze['author'].value_counts().head(20))

# Posts per day
print("\n posts per day :")
print(squeeze.groupby('date')['id'].count().to_string())

# Sentiment
print(squeeze['compound'].describe())

# Sentiment labels
squeeze['sentiment_label'] = squeeze['compound'].apply(
    lambda x: 'positive' if x >= 0.05 else ('negative' if x <= -0.05 else 'neutral')
)
print(squeeze['sentiment_label'].value_counts())

# Score
print(squeeze['score'].describe())

# Deleted/bots
suspicious = squeeze[squeeze['author'].isin(['[deleted]', 'AutoModerator'])]
print(f"[deleted] or AutoModerator posts: {len(suspicious)}")
```
